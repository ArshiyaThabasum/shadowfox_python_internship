```python
# Take height input from the user (in meters)
height = float(input("Enter height in meters: "))

# Take weight input from the user (in kilograms)
weight = float(input("Enter weight in kilograms: "))

# Calculate BMI
# Formula: BMI = weight / (height^2)
bmi = weight / (height ** 2)

# Display BMI value
print("BMI =", round(bmi, 2))

# Check BMI category using if-elif-else
if bmi >= 30:
    print("Obesity")

elif bmi >= 25:
    print("Overweight")

elif bmi >= 18.5:
    print("Normal")

else:
    print("Underweight")
```

```python
# List of cities in each country
Australia = ["Sydney", "Melbourne", "Brisbane", "Perth"]

UAE = ["Dubai", "Abu Dhabi", "Sharjah", "Ajman"]

India = ["Mumbai", "Bangalore", "Chennai", "Delhi"]

# Take two city names from the user
city1 = input("Enter the first city: ")
city2 = input("Enter the second city: ")

# Check if both cities belong to Australia
if city1 in Australia and city2 in Australia:
    print("Both cities are in Australia")

# Check if both cities belong to UAE
elif city1 in UAE and city2 in UAE:
    print("Both cities are in UAE")

# Check if both cities belong to India
elif city1 in India and city2 in India:
    print("Both cities are in India")

# If cities belong to different countries
else:
    print("They don't belong to the same country")
```
