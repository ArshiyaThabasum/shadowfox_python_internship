```python
# Store the radius of the pond in meters
radius = 84

# Store the value of pi
pi = 3.14

# Calculate the area of the pond
# Formula: Area = π × r²
area = pi * (radius ** 2)

# Print the area
print("Area of Pond =", area, "square meters")

# Water available per square meter
water_per_square_meter = 1.4

# Calculate total water in the pond
total_water = area * water_per_square_meter

# Print total water without decimal points
print("Total Water =", int(total_water), "liters")
```

```python
# Distance crossed in meters
distance = 490

# Time taken in minutes
time_minutes = 7

# Convert minutes into seconds
time_seconds = time_minutes * 60

# Calculate speed
# Formula: Speed = Distance / Time
speed = distance / time_seconds

# Print speed without decimal points
print("Speed =", int(speed), "m/s")
```
