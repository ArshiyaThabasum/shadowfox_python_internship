```python
# Create the Justice League list
justice_league = [
    "Superman",
    "Batman",
    "Wonder Woman",
    "Flash",
    "Aquaman",
    "Green Lantern"
]

# Display original list
print("Original List:")
print(justice_league)

# --------------------------------------------------
# 1. Calculate the number of members
# --------------------------------------------------
print("\n1. Number of Members:")
print(len(justice_league))

# --------------------------------------------------
# 2. Add Batgirl and Nightwing
# --------------------------------------------------
justice_league.extend(["Batgirl", "Nightwing"])

print("\n2. After Adding Batgirl and Nightwing:")
print(justice_league)

# --------------------------------------------------
# 3. Move Wonder Woman to the beginning
# --------------------------------------------------
justice_league.remove("Wonder Woman")
justice_league.insert(0, "Wonder Woman")

print("\n3. Wonder Woman Becomes Leader:")
print(justice_league)

# --------------------------------------------------
# 4. Separate Aquaman and Flash
# Move Superman between them
# --------------------------------------------------

# Remove Superman from current position
justice_league.remove("Superman")

# Find Aquaman's index
aquaman_index = justice_league.index("Aquaman")

# Insert Superman after Aquaman
justice_league.insert(aquaman_index + 1, "Superman")

print("\n4. After Separating Aquaman and Flash:")
print(justice_league)

# --------------------------------------------------
# 5. Replace the team with new members
# --------------------------------------------------
justice_league = [
    "Cyborg",
    "Shazam",
    "Hawkgirl",
    "Martian Manhunter",
    "Green Arrow"
]

print("\n5. New Justice League Team:")
print(justice_league)

# --------------------------------------------------
# 6. Sort alphabetically
# --------------------------------------------------
justice_league.sort()

print("\n6. Sorted Justice League:")
print(justice_league)

# Hero at index 0 becomes leader
print("\nNew Leader:")
print(justice_league[0])
```

```python
#Cyborg becomes the new leader because after sorting alphabetically, "Cyborg" appears at index 0.
```
