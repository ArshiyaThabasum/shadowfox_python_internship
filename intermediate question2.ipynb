```python
#INTERMEDIATE LEVEL question 2
# Hangman: Implement the wordguessing game with visualprogress and hints.

import random

# Dictionary containing words and hints
words = {
    "python": "A popular programming language",
    "elephant": "The largest land animal",
    "computer": "An electronic machine",
    "football": "A famous outdoor sport",
    "guitar": "A musical instrument"
}

# Randomly choose a word
word = random.choice(list(words.keys()))

# Get corresponding hint
hint = words[word]

# Store guessed letters
guessed_letters = []

# Number of attempts
attempts = 6

# Hangman stages
hangman = [
"""
 -----
 |   |
 |
 |
 |
 |
---------
""",
"""
 -----
 |   |
 |   O
 |
 |
 |
---------
""",
"""
 -----
 |   |
 |   O
 |   |
 |
 |
---------
""",
"""
 -----
 |   |
 |   O
 |  /|
 |
 |
---------
""",
"""
 -----
 |   |
 |   O
 |  /|\\
 |
 |
---------
""",
"""
 -----
 |   |
 |   O
 |  /|\\
 |  /
 |
---------
""",
"""
 -----
 |   |
 |   O
 |  /|\\
 |  / \\
 |
---------
"""
]

print("Welcome to Hangman!")
print("Hint:", hint)

while attempts > 0:

    # Display current progress
    display_word = ""

    for letter in word:
        if letter in guessed_letters:
            display_word += letter + " "
        else:
            display_word += "_ "

    print("\nWord:", display_word)

    # Check win condition
    if "_" not in display_word:
        print("Congratulations! You guessed the word:", word)
        break

    guess = input("Enter a letter: ").lower()

    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    if guess not in word:
        attempts -= 1

        print("Wrong guess!")
        print(hangman[6 - attempts])

        print("Attempts left:", attempts)

else:
    print("\nGame Over!")
    print("The word was:", word)
```
