```python
# Import random module to generate random numbers
import random

# Variables to count occurrences
count_6 = 0
count_1 = 0
two_sixes_in_a_row = 0

# Store previous roll
previous_roll = 0

# Roll the die 20 times
for i in range(20):

    # Generate a random number between 1 and 6
    current_roll = random.randint(1, 6)

    # Display each roll
    print("Roll", i + 1, "=", current_roll)

    # Count number of 6s
    if current_roll == 6:
        count_6 += 1

    # Count number of 1s
    if current_roll == 1:
        count_1 += 1

    # Check for two consecutive 6s
    if previous_roll == 6 and current_roll == 6:
        two_sixes_in_a_row += 1

    # Update previous roll
    previous_roll = current_roll

# Display statistics
print("\nStatistics")
print("Number of times 6 appeared:", count_6)
print("Number of times 1 appeared:", count_1)
print("Number of times two 6s occurred in a row:", two_sixes_in_a_row)
```

```python
# Variable to track completed jumping jacks
completed_jumping_jacks = 0

# Loop 10 times (10 sets of 10 jumping jacks = 100)
for i in range(10):

    # Complete 10 jumping jacks
    completed_jumping_jacks += 10

    # Check if workout is completed
    if completed_jumping_jacks == 100:
        print("Congratulations! You completed the workout.")
        break

    # Ask if the user is tired
    tired = input("Are you tired? (yes/y or no/n): ")

    # If user is tired
    if tired.lower() == "yes" or tired.lower() == "y":

        # Ask whether to skip remaining sets
        skip = input("Do you want to skip the remaining sets? (yes/y or no/n): ")

        # If user wants to skip
        if skip.lower() == "yes" or skip.lower() == "y":
            print("You completed a total of", completed_jumping_jacks, "jumping jacks.")
            break

    # Calculate remaining jumping jacks
    remaining = 100 - completed_jumping_jacks

    # Display remaining jumping jacks
    print(remaining, "jumping jacks remaining.")
```
